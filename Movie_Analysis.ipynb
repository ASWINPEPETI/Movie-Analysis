```python
df_UK= df.copy()
df_UK["country"]= df_UK["country"].str.split(",")
df_UK=df_UK.explode("country")
```

```python
df_UK=df_UK[df_UK["country"]=="United States"]
```

```python
df_UK=df_UK.groupby("release_year").size().reset_index()
```

```python
df_UK
```

```python
df_UK['release_year'] = df_UK["release_year"].astype('category')
```
